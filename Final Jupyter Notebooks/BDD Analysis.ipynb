```python

import sys
print(sys.version)
import numpy as np
print(np.__version__)
import pandas as pd
print(pd.__version__)
import matplotlib.pyplot as plt
import geocoder
from geopy.geocoders import Nominatim
%matplotlib inline
import seaborn as sns
import plotly.graph_objects as go
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
# Load PIRUS Data
pirus = pd.read_csv('../data/clean_data/pirus_deep_clean_Final.csv')
pirus.set_index(pirus['subject_id'], inplace = True)
pirus.drop('subject_id', axis = 1, inplace = True)

# Load Brain Drain Data (BDD)
bdd = pd.read_csv('../data/clean_data/bdd_deep_clean_40221.csv')

state_codes = pd.read_csv('../data/clean_data/state_codes.json')

# mask = (pirus['year'] >= 2010) & (pirus['year'] <= 2012)
# pirus = pirus.loc[mask]

# m = (bdd['year'] >= 2010) & (bdd['year'] <= 2012)
# bdd = bdd.loc[m]

mask = pirus['year'] == 2017
pirus = pirus.loc[mask]

m = bdd['year'] == 2017
bdd = bdd.loc[m]
```

```python
fig = go.Figure()

fig.add_trace(
    go.Choropleth(
        locations=bdd['state_code'],
        z=bdd['abs_net_drain'].astype(float),
        locationmode='USA-states',
        colorscale='BrBG',
        autocolorscale=False,
        marker_line_color='white',
        colorbar_title="Net Drain"
    ))
        
fig.add_trace(
    go.Scattergeo(
        lon = pirus['loc_plot1_long'],
        lat = pirus['loc_plot1_lat'],
        mode = 'markers',
        marker = dict(
            size = 6,
            opacity = 1,
            reversescale = True,
            autocolorscale = False,
            symbol = 'x',
            line = dict(
                width=1,
                color='#800000'
            ),
    )))

fig.update_layout(
    title_text='Net Brain Drain and Plot Locations in 2017',
    geo = dict(
        scope='usa',
        projection=go.layout.geo.Projection(type = 'albers usa'),
        showlakes=True,
        lakecolor='rgb(255, 255, 255)'),
)

fig.show()
```
