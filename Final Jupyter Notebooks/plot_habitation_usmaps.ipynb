```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geocoder
from geopy.geocoders import Nominatim
%matplotlib inline
import seaborn as sns
import plotly.graph_objects as go
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

# Load PIRUS Data
pirus = pd.read_csv('../data/clean_data/pirus_deep_clean_Final.csv')
pirus.set_index(pirus['subject_id'], inplace = True)
pirus.drop('subject_id', axis = 1, inplace = True)
```

```python
#Build Scatter plot of plot locations ince 1981

fig = go.Figure()

fig.add_trace(
    go.Scattergeo(
        lon = pirus['loc_plot1_long'],
        lat = pirus['loc_plot1_lat'],
        mode = 'markers',
        marker = dict(
            size = 8,
            opacity = 1,
            reversescale = True,
            autocolorscale = False,
            color = '#ff7f0f',
            symbol = 'x',
            line = dict(
                width=1,
                color='rgba(102, 102, 102)'
            ),
    )))


fig.update_layout(
    title_text='Where Terrorist Plot or Activity Was Centered Since 1981',
    geo = dict(
        scope='usa',
        projection=go.layout.geo.Projection(type = 'albers usa'),
        showlakes=True,
        lakecolor='rgb(255, 255, 255)'),
)

fig.show()
```

```python
#Build Scatter plot of habitation locations ince 1981

fig = go.Figure()

fig.add_trace(
    go.Scattergeo(
        lon = pirus['loc_habit1_long'],
        lat = pirus['loc_habit1_lat'],
        mode = 'markers',
        marker = dict(
            size = 8,
            opacity = 1,
            reversescale = True,
            autocolorscale = False,
            symbol = 'circle',
            color = '#1f77b4',
            line = dict(
                width=1,
                color='rgba(102, 102, 102)'
            ),
    )))

fig.update_layout(
    title_text='Where Radicalized Individuals Lived Since 1981',
    geo = dict(
        scope='usa',
        projection=go.layout.geo.Projection(type = 'albers usa'),
        showlakes=True,
        lakecolor='rgb(255, 255, 255)'),
)

fig.show()
```
